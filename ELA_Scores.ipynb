```python
#dependencies
import csv
import pandas as pd
import numpy as np
```

```python
#create db from Resources
ela_csv = "Resources/School_ELA.csv"
ela_df = pd.read_csv(ela_csv)
ela_df.head()
```

```python
#Check size, number of unique DBN's
ela_df.shape
```

```python
ela_df['DBN'].nunique()
```

```python
#That is a lot but only joining on the same DBN will filter out the ones not in other db's
#Only want certain columns
ela_df = ela_df[['DBN', 'School Name', 'Year', 'Mean Scale Score']]
df_filtered = ela_df[ela_df['Year'] == 2015]
df_filtered.head()
df_filtered.shape
```

```python
#there are multiple entries for every year. Cool!
#the mode for scores is a placeholder value s, remove from df
for index, row in df_filtered.iterrows():
    if row['Mean Scale Score'] == 's':
        df_filtered.drop(index, inplace=True)
df_filtered.head()
```

```python
#To only get one result per DBN, groupby
df_filtered['Mean Scale Score'] = pd.to_numeric(df_filtered['Mean Scale Score'])
grouped_df = df_filtered.groupby(['DBN', 'School Name','Year']).mean()
display(grouped_df)
```

```python
#This is still more DBN's than our main df. Joining where they are = should fix
grouped_df.rename(columns={"DBN":"School ID", "Mean Scale Score": "Average ELA Score"})
```
