```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "scores.csv"
df = pd.read_csv(csv_file)
df.head()
```

```python
locations = df[['School ID', 'School Name', 'Borough', 'Building Code', 'Street Address', 'City', 'State', 'Zip Code', 'Latitude',
              'Longitude', 'Phone Number']].copy()
#locations = df.drop(columns=['Start Time','End Time', 'Student Enrollment', 'Percent White', 'Percent Black', 
#                             'Percent Hispanic', 'Percent Asian', 'Average Score (SAT Math)', 'Average Score (SAT Reading)','Average Score (SAT Writing)',
#                            'Percent Tested'])
locations = locations.dropna()
locations = locations.drop_duplicates(keep='last')
locations.head()
```

```python
students = df.drop(columns=['Start Time', 'End Time','School Name', 'Borough', 'Building Code', 'Street Address', 'City', 'State', 'Zip Code', 'Latitude', 'Longitude', 
                            'Phone Number', 'Average Score (SAT Math)', 'Average Score (SAT Reading)','Average Score (SAT Writing)',
                             'Percent Tested'])
students = students.rename(columns={'Student Enrollment': 'Enrollment', 'Percent White': 'White', 'Percent Black': 'Black', 
                                   'Percent Hispanic': 'Hispanic', 'Percent Asian': 'Asian'})
students = students.dropna()
students = students.drop_duplicates(keep='last')
students.head()
```

```python
SAT = df.drop(columns=['Start Time', 'End Time','School Name', 'Borough', 'Building Code', 'Street Address', 'City', 'State', 'Zip Code', 'Latitude', 
                  'Longitude', 'Phone Number','Start Time','End Time', 'Student Enrollment', 'Percent White', 'Percent Black', 
                   'Percent Hispanic', 'Percent Asian'])
SAT = SAT.rename(columns={'Average Score (SAT Math)': 'Avg Score Math', 'Average Score (SAT Reading)': 'Avg Score Reading',
                         'Average Score (SAT Writing)': 'Avg Score Writing', 'Percent Tested': '% Tested'})
SAT = SAT.dropna()
SAT = SAT.drop_duplicates(keep='last')
SAT.head()
```

```python
df_cleaned = df.rename(columns={'Student Enrollment': 'Enrollment', 'Percent White': 'White', 'Percent Black': 'Black', 
                                   'Percent Hispanic': 'Hispanic', 'Percent Asian': 'Asian','Average Score (SAT Math)': 'Avg Score Math', 'Average Score (SAT Reading)': 'Avg Score Reading',
                         'Average Score (SAT Writing)': 'Avg Score Writing', 'Percent Tested': '% Tested'})
df_cleaned = df_cleaned.dropna()
df_cleaned = df_cleaned.drop_duplicates()
df_cleaned.head()
```
