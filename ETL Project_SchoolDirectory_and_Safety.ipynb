School Safety

```python
safety_csv = "Resources/SchoolSafety.csv"
safety_db = pd.read_csv(safety_csv)
safety_db.head()
```

```python
safety_db.columns
```

```python
safety_db.rename(columns={'DBN': 'school id'}, inplace = True)
safety_db = safety_db[['school id', 'Building Code','Location Name', 'Major N', 'Oth N',
       'NoCrim N', 'Prop N', 'Vio N']].copy()
safety_db.head()
```

```python
safety_db.shape
```

```python
#drop rows with no reported crimes 
safety_db.dropna(subset=['Major N', 'Oth N', 'NoCrim N', 'Prop N', 'Vio N'])
safety_db.shape
```

```python
safety_db.head()
```

```python
safety_db['total_crime'] = safety_db['Major N']+ safety_db['Oth N']+safety_db['NoCrim N']+safety_db['Prop N']+safety_db['Vio N']
safety_db.head()
```
