```python
#set up Socrata client
client = Socrata("data.cityofnewyork.us", None)

#get results
results = client.get("3y5p-x48f", limit=843113)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)

#show reults
results_df
```

```python
#delete columns
school_df=results_df[['school','date','schoolyear','enrolled','present','absent']]
school_df
```

```python
#convert date to datetime
school_df['date']=pd.to_datetime(school_df['date'])

#convert to integers
school_df['enrolled']=school_df['enrolled'].astype(int)
school_df['present']=school_df['present'].astype(int)
school_df['absent']=school_df['absent'].astype(int)

#rename school to school_id which is the agreed column name that will join the group db together 
school_df.rename(columns={'school':'school_id'},inplace=True)
```

```python
#verify
school_df.head()
```

```python
#organized by school_id and school year
school_attendance_df=school_df.groupby(['school_id']).mean()
school_attendance_df.head()
```

```python
#add percentage column
school_attendance_df['attendance_prc']=round(school_attendance_df['present']/school_attendance_df['enrolled']*100,2)
```

```python
school_attendance_df.rename(columns={'enrolled':'enrolled_2015_2018','present':'present_2015_2018','absent':'absent_2015_2018'},inplace=True)
school_attendance_df.head()
```

```python
#create db from Resources
ela_csv = "Resources/School_ELA.csv"
ela_df = pd.read_csv(ela_csv)
ela_df.head()
```

```python
#Check size, number of unique DBN's
ela_df.shape
```

```python
ela_df['DBN'].nunique()
```

```python
#That is a lot but only joining on the same DBN will filter out the ones not in other db's
#Only want certain columns
ela_df = ela_df[['DBN', 'School Name', 'Year', 'Mean Scale Score']]
df_filtered = ela_df[ela_df['Year'] == 2015]
df_filtered.head()
df_filtered.shape
```

```python
#there are multiple entries for every year. Cool!
#the mode for scores is a placeholder value s, remove from df
for index, row in df_filtered.iterrows():
    if row['Mean Scale Score'] == 's':
        df_filtered.drop(index, inplace=True)
df_filtered.head()
```

```python
df_filtered.rename(columns={'DBN':'school id'},inplace=True)
```

```python
#To only get one result per DBN, groupby
df_filtered['Mean Scale Score'] = pd.to_numeric(df_filtered['Mean Scale Score'])
grouped_df = df_filtered.groupby(['school id', 'School Name','Year']).mean()
display(grouped_df)
```

```python
#This is still more DBN's than our main df. Joining where they are = should fix
grouped_df.rename(columns={"DBN":"school id", "Mean Scale Score": "Average ELA Score"})
grouped_df.head()
```

```python
#changed so just school id is index
grouped_df['School Name']=grouped_df.index.get_level_values('School Name')
grouped_df['Year']=grouped_df.index.get_level_values('Year')
grouped_df['school id']=grouped_df.index.get_level_values('school id')
grouped_df.set_index('school id',inplace=True)
grouped_df
```
